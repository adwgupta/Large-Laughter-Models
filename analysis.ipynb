```python
import os
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
import os
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# Define the function to process each Excel file
def process_excel(file_path):
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_path, engine='openpyxl')
    
    # Calculate verbosity for each entry in "Set 1" and "Set 2"
    df['Verbosity Set 1'] = df['Set 1'].apply(lambda x: len(word_tokenize(str(x))))
    df['Verbosity Set 2'] = df['Set 2'].apply(lambda x: len(word_tokenize(str(x))))
    
    # Apply the function to check profanity and filter out inappropriate content
    df['Profanity Set 1'] = df.apply(lambda row: check_profanity(str(row['Setup']) + " " + str(row['Set 1'])), axis=1)
    df['Profanity Set 2'] = df.apply(lambda row: check_profanity(str(row['Setup']) + " " + str(row['Set 2'])), axis=1)
    
    return df

# Initialize dictionary to store results
results = {}

# Iterate over each Excel file within the "FT Jokes" folder
folder_path = "FT Jokes"
for file_name in os.listdir(folder_path):
    if file_name.endswith('.xlsx') and not file_name.startswith('$') and not file_name.startswith('~$'):
        # Process the Excel file only if it doesn't start with '$' or '~$'
        file_path = os.path.join(folder_path, file_name)
        df = process_excel(file_path)
        results[file_name] = df

for file_name, df in results.items():
    print(f"File: {file_name}")
    print(df)

```
